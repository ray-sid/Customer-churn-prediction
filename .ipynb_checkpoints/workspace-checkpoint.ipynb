```python
import pandas as pd
import numpy as np
df = pd.read_csv("Telco-Customer-Churn.csv")
```

```python
df
```

```python
df.dtypes
```

```python
df.drop("customerID",axis = "columns",inplace = True)
```

```python
df.dtypes
```

```python
for col in df:
    if(df[col].dtypes == 'object'):
        print(f'{col}: {df[col].unique()}')
```

```python
yes_no = ['Partner','Dependents','PaperlessBilling','Churn','PhoneService']

for i in yes_no:
    df[i].replace({'Yes':1,'No':0},inplace = True)
```

```python
for col in df:
    print(f'{col}: {df[col].unique()}')
```

```python
for i in ["gender"]:
    df[i].replace({'Male':1,'Female':0},inplace = True)
```

```python
def uniq(a):
    for col in a:
        print(f'{col}: {a[col].unique()}')
```

```python
uniq(df)
```

```python
df.replace('No internet service','No',inplace = True)
df.replace('No phone service','No',inplace = True)
```

```python
uniq(df)
```

```python
yes_no = ['MultipleLines','OnlineSecurity','OnlineBackup','DeviceProtection','TechSupport','StreamingTV','StreamingMovies']

for i in yes_no:
    df[i].replace({'Yes':1,'No':0},inplace = True)
```

```python
uniq(df)
```

```python
df = pd.get_dummies(data = df,columns = ['InternetService','Contract','PaymentMethod'])
```

```python
uniq(df)
```

```python
df.dtypes
```

```python
'''s = df['TotalCharges']
s[488]'''

df['TotalCharges'].replace(" ", np.nan, inplace=True)
```

```python
df.dropna(subset=['TotalCharges'], inplace=True)
```

```python
df['TotalCharges'] = pd.to_numeric(df['TotalCharges'])
```

```python
uniq(df)
```

```python
#scaling - converting the values in the range of 0 to 1

s_cols = ['tenure','MonthlyCharges','TotalCharges']
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
df[s_cols] = scaler.fit_transform(df[s_cols])
```

```python
uniq(df)
```

```python
df.sample(2)
```

```python
x = df.drop('Churn',axis = 'columns')
y = df['Churn']
```

```python
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest = train_test_split(x,y,test_size = 0.2,random_state = 5)
```

```python
xtrain.shape
```

```python
#creating a neural network
import tensorflow as tf
from tensorflow import keras
model = keras.Sequential([
    keras.layers.Dense(26,input_shape = (26,),activation = 'relu'), #input layer, neurons as no of columns ie 26
    keras.layers.Dense(15,activation = 'relu'), #no need for input_shape cause it knows the derived shape
    keras.layers.Dense(1,activation = 'sigmoid'), 
])
```

```python
model.compile(optimizer = 'adam',
             loss = 'binary_crossentropy',
             metrics = ['accuracy'] )
```

```python
model.fit(xtrain,ytrain,epochs = 100)
```

```python
model.evaluate(xtest,ytest)
```

```python
ypredict = model.predict(xtest)
ypredict[:5]
```
