```python
import pandas as pd
df = pd.read_csv("telco_churn_data.csv")
```

```python
df.dtypes
```

```python
df.drop("Customer ID",axis = "columns",inplace = True)
```

```python
df.dtypes
```

```python
for col in df:
    if(df[col].dtypes == 'object'):
        print(f'{col}: {df[col].unique()}')
```

```python
yes_no = ['Referred a Friend','Phone Service','Multiple Lines','Internet Service','Online Security','Online Backup','Device Protection Plan','Premium Tech Support','Streaming TV','Streaming Movies','Streaming Music','Unlimited Data','Paperless Billing','Under 30','Senior Citizen','Married','Dependents',]

for i in yes_no:
    df[i].replace({'Yes':1,'No':0},inplace = True)
```

```python
for col in df:
    print(f'{col}: {df[col].unique()}')
```

```python
for i in ["Gender"]:
    df[i].replace({'Male':1,'Female':0},inplace = True)
```

```python
for col in df:
    print(f'{col}: {df[col].unique()}')
```
